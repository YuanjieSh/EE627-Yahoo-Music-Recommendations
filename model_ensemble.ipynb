```python
import pandas as pd
import numpy as np
```

```python
# min\\(s1*w1 + s2*w2) - y(truth)\\^2  (s is the submission, w is corresponding weight)
# Want to get W(Matrix of w)
# Since minimize (SW - y)T (SW - y) -> W = (ST * S)^-1 * ST * y
# ST * y = K * [(2*correct_rate ) -1 ]
```

```python
# First try, ensemble two submissions
sub_1 = pd.read_csv('EE627_final_ver8.csv')
sub_2 = pd.read_csv('EE627_final_ver10.csv') 
```

```python
# Replace 0 by -1
sub_1.Predictor.replace(0,-1,inplace= True)
sub_2.Predictor.replace(0,-1,inplace= True)
```

```python
sub_1
```

```python
s_1 = sub_1.Predictor.tolist()
s_2 = sub_2.Predictor.tolist()
```

```python
s_2
```

```python
from numpy import*
```

```python
S = mat([s_1,s_2])
```

```python
S = np.transpose(S)
S
```

```python
# score for two model
c_1 = 0.81183
c_2 = 0.68225
c_mat = np.transpose(mat([c_1,c_2]))
c_mat
```

```python
# ST * y 
k = 120000
STy = k* (2*c_mat - 1)
STy
```

```python
STS = np.transpose(S) * S
STS_inv = STS.I
STS_inv
```

```python
np.set_printoptions(suppress=False)
```

```python
W = STS_inv * STy
W
```

```python
new_ratings = S * W
new_ratings
```

```python
new_ratings = new_ratings.T
```

```python
new_ratings = np.array(new_ratings).reshape((20000,6)).tolist()
new_ratings
```

```python
sub_1
```

```python
# Function to create trackID and predicton score table

def prediction_table(grades_table):
    prediction_table = [0 for i in range(len(grades_table)*6)]
    for i in range(len(grades_table)):
        x = np.array(grades_table[i])
        x = x.argsort()

        for k in range(6):
#             prediction_table[i * 6 + k][0] = str(test_set[i][0]) + '_' + str(test_set[i][k + 2])
            if k<3:
                prediction_table[i * 6 + x[k]] = 0
            else:
                prediction_table[i * 6 + x[k]] = 1
      
    return prediction_table
```

```python
new_ratings_table = prediction_table(new_ratings)
new_ratings_table
```

```python
sub_2.Predictor = new_ratings_table
```
